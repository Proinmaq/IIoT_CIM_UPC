```python
from bigml.api import BigML
```

```python
api = BigML ('infoproinmaq','76f2d4ac92608cef58641218771e41f94449e159')
```

```python
proyecto = api.create_project({'name':'Sesion 7 UPC CIM'})
```

```python
api = BigML ('infoproinmaq','76f2d4ac92608cef58641218771e41f94449e159', project = proyecto['resource'])
```

```python
source = api.create_source('https://cleverdata.io/csv/telecom.csv')
api.ok(source)
```

```python
dataset = api.create_dataset(source)
api.ok(dataset)
```

```python
train_dataset_ori = api.create_dataset(dataset,{"name":"telecom 80%","sample_rate":0.8,"seed":"mi semilla"})
test_dataset_ori = api.create_dataset(dataset,{"name":"telecom 20%","sample_rate":0.8,"seed":"mi semilla","out_of_bag": True})
api.ok(train_dataset_ori)
api.ok(test_dataset_ori)
```

```python
def crear_modelos_evaluaciones(conjunto_entrenamiento,conjunto_test):
  arbol = api.create_model(conjunto_entrenamiento)
  api.ok(arbol)
  evaluacion_arbol = api.create_evaluation(arbol,conjunto_test)

  ensemble = api.create_ensemble(conjunto_entrenamiento)
  api.ok(ensemble)
  evaluacion_ensemble = api.create_evaluation(ensemble,conjunto_test)

  RN = api.create_deepnet(conjunto_entrenamiento)
  api.ok(RN)
  evaluacion_RN = api.create_evaluation(RN,conjunto_test)
```

```python
crear_modelos_evaluaciones(train_dataset_ori, test_dataset_ori)
```

```python
detector_anomalias = api.create_anomaly(train_dataset_ori, {"excluded_fields":["Churn"]})
api.ok(detector_anomalias)
```

```python
batch_anomaly_train = api.create_batch_anomaly_score(detector_anomalias, train_dataset_ori, {"all_fields": True, "output_dataset": True})
api.ok(batch_anomaly_train)

batch_anomaly_test = api.create_batch_anomaly_score(detector_anomalias, test_dataset_ori, {"all_fields": True, "output_dataset": True})
api.ok(batch_anomaly_test)
```

```python
dataset_con_score = batch_anomaly_train['object']['output_dataset_resource']
cambio_objetivo = api.update_dataset(dataset_con_score, {"objective_field":{"id":"000013"}})
```

```python
crear_modelos_evaluaciones(batch_anomaly_train['object']['output_dataset_resource'], batch_anomaly_test['object']['output_dataset_resource'])
```
